```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
X=6*np.random.rand(100,1)-3
y=0.5*X**2+1.58*X+2+np.random.rand(100,1)
plt.scatter(X,y,color='r')
plt.xlabel('X datasets')
plt.ylabel('y datasets')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
regression1= LinearRegression()
```

```python
regression1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression1.predict(X_test))
score
```

```python
plt.plot(X_train,regression1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
plt.xlabel('X datasets')
plt.ylabel('y datasets')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.scatter(X_train,regression.predict(X_train_poly))
plt.scatter(X_train,y_train)
```

Polynomial with degree of 3

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

Prediction of new dataset

```python
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
```

```python
y_new=regression.predict(X_new_poly)
plt.plot(X_new,y_new,'r-',linewidth=2,label="Predictions")
plt.plot(X_train,y_train,'b.',label='Training Points')
plt.plot(X_test,y_test,'g.',label='Testing Points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
